```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pde
```

```python
S = 100
r = 0.05
q = 0.02
sigma = 0.2
M = 100
N = 100
T = 2
K = 30
otype = "put"
is_am = "True"

t_values, s_values, f, price = pde.pde(S, r, q, sigma, M, N, T, K, otype, is_am)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection = '3d')

S, T = np.meshgrid(t_values, s_values)

surf = ax.plot_surface(S, T, f.T, cmap = 'viridis', edgecolor = 'k')

ax.set_xlabel('Stock Price')
ax.set_ylabel('Time')
ax.set_zlabel('Option Price')
ax.set_title('American Put Option')

plt.show()
```
